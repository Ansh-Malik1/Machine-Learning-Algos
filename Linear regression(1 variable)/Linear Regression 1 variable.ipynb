```python
x = data.iloc[:,0].values  # R&D spend(independent variable)
print(x)
```

```python
y = data.iloc[:,-1].values  # Profit(dependable variable)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=1/10,random_state=0)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
regression_model.fit(x_train.reshape(-1,1),y_train.reshape(-1,1))
```

```python
y_predicted = regression_model.predict(x_test.reshape(-1,1))
y_predicted
```

```python
plt.plot(x_test,y_predicted,'r-*')
plt.xlabel('R&D spend')
plt.ylabel('Proift')
plt.title('Graph of profit and R&D spend')
```

```python
print(regression_model.coef_)
```

```python
print(regression_model.intercept_)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predicted) 
```
